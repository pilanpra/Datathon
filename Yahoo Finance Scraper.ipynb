```python
import pandas as pd
import urllib.request
import json
from datetime import datetime as dt

# Function to fetch ESG data for a specific stock symbol
def fetch_esg_data(symbol):
    url = f"https://query2.finance.yahoo.com/v1/finance/esgChart?symbol={symbol}"
    connection = urllib.request.urlopen(url)
    data = connection.read()
    data_2 = json.loads(data)

    # Extract the ESG data series
    try:
        Formatdata = data_2["esgChart"]["result"][0]["symbolSeries"]
        Formatdata_2 = pd.DataFrame(Formatdata)
        Formatdata_2["timestamp"] = pd.to_datetime(Formatdata_2["timestamp"], unit="s")
        Formatdata_2["symbol"] = symbol  # Add symbol column for identification
        return Formatdata_2
    except KeyError as e:
        print(f"No ESG data found for {symbol}: {e}")
        return pd.DataFrame()

# List of stock symbols to fetch data for
symbols = djia['Tickers']

# Fetch ESG data for each symbol and combine into a single DataFrame
all_data = pd.DataFrame()
for symbol in symbols:
    print(f"Fetching data for {symbol}...")
    esg_data = fetch_esg_data(symbol)
    all_data = pd.concat([all_data, esg_data], ignore_index=True)

# Save the combined data to a CSV file
all_data.to_csv("ESG_Historical_Data.csv", index=False)
print("Data saved to ESG_Historical_Data.csv")
```

```python
import pandas as pd

# Assuming your data is in a DataFrame called 'df'
# If it's not, you can load it from a CSV file like this:
df = pd.read_csv('ESG_Historical_Data.csv')

# Convert the timestamp column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d')

# Extract the year from the timestamp
df['year'] = df['timestamp'].dt.year

# Group by year and symbol, then calculate the mean for numeric columns
aggregated_data = df.groupby(['year', 'symbol']).agg({
    'esgScore': 'mean',
    'governanceScore': 'mean',
    'environmentScore': 'mean',
    'socialScore': 'mean'
}).reset_index()

# Round the numeric columns to two decimal places
numeric_columns = ['esgScore', 'governanceScore', 'environmentScore', 'socialScore']
aggregated_data[numeric_columns] = aggregated_data[numeric_columns].round(2)

# Sort the data by symbol and year
aggregated_data = aggregated_data.sort_values(['symbol', 'year'])

# Display the aggregated data
print(aggregated_data)

# Optionally, save the aggregated data to a CSV file
aggregated_data.to_csv('aggregated_esg_data.csv', index=False)
```

```python
import yfinance as yf
import pandas as pd

# List of stock symbols to fetch data for
symbols =  df['symbol']

# Initialize a DataFrame to store all data
all_data = []

# Function to fetch ESG scores using yfinance
def fetch_esg_data(symbol):
    try:
        # Fetch stock ticker data
        ticker = yf.Ticker(symbol)

        # Extract sustainability (ESG) data
        esg_data = ticker.sustainability
        
        # Add symbol to data for identification
        if esg_data is not None:
            esg_data["symbol"] = symbol
            return esg_data
        else:
            print(f"No ESG data available for {symbol}.")
            return None
    except Exception as e:
        print(f"Failed to fetch data for {symbol}: {e}")
        return None

# Fetch ESG data for all symbols
for symbol in symbols:
    print(f"Fetching data for {symbol}...")
    esg_data = fetch_esg_data(symbol)
    if esg_data is not None:
        all_data.append(esg_data)

# Combine all ESG data into a single DataFrame
if all_data:
    all_data_df = pd.concat(all_data)
    output_file = "ESG_Data_Yahoo_Finance_Tickers.csv"
    all_data_df.to_csv(output_file, index=True)
    print(f"ESG data saved to {output_file}")
else:
    print("No ESG data was available for any of the symbols.")
```
